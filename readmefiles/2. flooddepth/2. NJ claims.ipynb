```python
import pandas as pd
pd.set_option('display.max_columns', None)

file_path = 'NJ.csv'

df = pd.read_csv(file_path)

df
```

```python
pd.set_option('display.max_columns', None)
```

```python
filtered_df = df[df['disasterNumber'] == 4614]
filtered_df
```

```python
filtered_df.groupby('damagedZipCode').size()
```

```python
# How many people who called FEMA are flood insured?
filtered_df.groupby('floodInsurance').size()
```

```python
# Grouping the DataFrame by 'damagedZipCode' and calculating the average of 'waterLevel'
grouped_df = filtered_df.groupby('damagedZipCode')['waterLevel'].mean().reset_index()

# Creating a new DataFrame with columns 'Zipcode' and 'Water depth'
new_df = pd.DataFrame(columns=['Zipcode', 'Water depth'])

# Assigning unique zip codes from 'damagedZipCode' column to 'Zipcode' column in new_df
new_df['Zipcode'] = grouped_df['damagedZipCode'].unique()

# Assigning average water levels to 'Water depth' column in new_df
new_df['Water depth'] = grouped_df['waterLevel']

# Displaying the new DataFrame
print(new_df)
```

```python
# Loop through the Zipcode column in new_df
for i in range(len(new_df)):
    # Check if the Zipcode is four digits
    if len(str(new_df.loc[i, 'Zipcode'])) == 4:
        # Add "0" in front of the string
        new_df.loc[i, 'Zipcode'] = '0' + str(new_df.loc[i, 'Zipcode'])

# Print the updated new_df DataFrame
print(new_df)
```

```python
# Save new_df to a CSV file
new_df.to_csv('new_df.csv', index=False)
```
